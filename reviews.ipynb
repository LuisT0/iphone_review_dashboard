```python
import pandas as pd
import kagglehub
import re
import numpy as np
```

```python
#descargar dataset

path = kagglehub.dataset_download("mrmars1010/iphone-customer-reviews-nlp")

print("Path to dataset files:", path)
```

```python
#importar desde ruta relativa local

iphone_df = pd.read_csv("~/Documents/DataSets/iphone.csv")

iphone_df.head(10)
```

```python
#buscar asin √∫nicos

asin_unique = iphone_df['productAsin'].value_counts().unique

asin_unique
```

```python
#diccionario de modelos de iphone

iphone_dict = {'B09G9D8KRQ': 'iPhone 13',
               'B0BN72MLT2': 'iPhone 14',
               'B09G9BL5CP': 'iPhone 13',
               'B0BDK8LKPJ': 'iPhone 14',
               'B0CHX1W1XY': 'iPhone 15',
               'B09P82T3PZ': 'iPhone 13',
               'B09G9J5JZX': 'iPhone 13'}

#a√±adir columna product name vac√≠a

iphone_df['product_name'] = None 

#bucle para reemplazar por nombres comerciales

for asin, model in iphone_dict.items():
    mask = iphone_df['productAsin'] == asin
    iphone_df.loc[mask, 'product_name'] = model

columns_order = ['product_name', 'productAsin', 'variant', 'country', 'date', 'isVerified', 'ratingScore', 'reviewTitle', 'reviewDescription', 'reviewUrl', 'reviewedIn', 'variantAsin']
iphone_df = iphone_df[columns_order]
```

```python
#comprobar cambios

iphone_df.head(3)
```

```python
#eliminar columnas que no aportan valor

iphone_df = iphone_df.drop(columns=['isVerified', 'reviewTitle', 'reviewUrl', 'reviewedIn', 'variantAsin'], axis=0)

#comprobar cambios

iphone_df.head(3)
```

```python
#revisar por valores √∫nicos en variant
iphone_df['variant'].value_counts().unique
```

```python
#crear copia de la columna para vectorizar

iphone_df['variant_clean'] = iphone_df['variant']
```

```python
#estandarizar val√≥res an√≥malos en variant_clean

iphone_df['variant_clean'] = iphone_df['variant_clean'].replace(regex=['Colour:', 'Service Provider:'], value=['Color:', 'Color:'])
```

```python
#reordenar strings que no siguen el orden deseado

iphone_df['variant_clean'] = iphone_df['variant_clean'].str.replace(
    r'^Size:\s*(?P<size>.+?)\s*Color:\s*(?P<color>.+)$',
    r'Color: \g<color> Size: \g<size>',
    regex=True
)
```

```python
#estandarizar valores que incluyan un carrier antes de color

iphone_df['variant_clean'] = iphone_df['variant_clean'].replace(regex=['GSM CarriersColor:', 'VerizonColor:',
                                                                       'AT&TColor:', 'T-MobileColor:',
                                                                       'UnlockedColor:'], value=[' ', ' ', ' ', ' ', 'Color:'])
```

```python
#extraer color y storage para crear columnas independientes

iphone_df[['color', 'storage']] = iphone_df['variant_clean'].str.extract(
    r'Color:\s*(.+?)\s*Size:\s*(.+)'
)
```

```python
iphone_df['color'].value_counts().unique
```

```python
#limpiar valores √∫nicos

iphone_df['color'] = (iphone_df['color'].str.replace('Color:', '', regex=False).str.strip())

#comprobar cambios
iphone_df['color'].value_counts().unique
```

```python
#Estandarizar colores
iphone_df['color'] = iphone_df['color'].replace(regex=['blue', 'green','starlight'], value=['Blue', 'Green','Starlight'])

iphone_df['color'] = iphone_df['color'].str.replace(
    r'\(Product\)\s*Red',
    r'(PRODUCT) RED',
    regex=True
)

iphone_df['color'] = iphone_df['color'].str.replace(
    r'\(Product\)\s*RED',
    r'(PRODUCT) RED',
    regex=True
)
```

```python
#comprobar cambios
iphone_df['color'].value_counts().unique
```

```python
#visualizar valores √∫nicos en storage
iphone_df['storage'].value_counts().unique
```

```python
#estandarizar valores en storage
iphone_df['storage'] = iphone_df['storage'].str.replace(
    r'(\d+)\s*GB',
    r'\1 GB',
    regex=True
)
```

```python
#comprobar cambios
iphone_df['storage'].value_counts().unique
```

```python
#convertir storage a num√©rico
iphone_df['storage_gb'] = (
    iphone_df['storage']
    .str.extract(r'(\d+)', expand=False)
    .astype('Int64')
)

#comprobar cambios

print(iphone_df['storage_gb'].value_counts())
print(iphone_df['storage_gb'].dtype)
```

```python
iphone_df['ratingScore'].value_counts().unique
```

```python
# crear columna para agrupar el sentimiento de los ratings

conditions = [
    iphone_df['ratingScore'] <= 2,
    iphone_df['ratingScore'] == 3,
    iphone_df['ratingScore'] >= 4
]

choices = ['Negativo', 'Neutro', 'Positivo']

iphone_df['sentiment'] = np.select(conditions, choices, default='Sin clasificar')

iphone_df['sentiment'].value_counts()
```

```python
#reordenar columnas y desechar las que no son relevantes

iphone_df = iphone_df.drop(columns=['variant', 'variant_clean', 'storage'], axis=0)

columns_order = ['product_name', 'color', 'storage_gb', 'country', 'date', 'ratingScore', 'sentiment', 'reviewDescription']
iphone_df = iphone_df[columns_order]
```

```python
#convertir date a fecha

iphone_df['date'] = pd.to_datetime(iphone_df['date'], dayfirst=True, errors='coerce')
```
